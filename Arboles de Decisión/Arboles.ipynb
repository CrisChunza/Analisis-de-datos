```python
#Tratamiento de los datos,
vino=vino[vino["fixed acidity"]<=12]
vino=vino[vino["volatile acidity"]<=1.0]
vino=vino[vino["citric acid"]<=0.8]
vino=vino[vino["residual sugar"]<=10]
vino=vino[vino["chlorides"]<=0.45]
vino=vino[vino["free sulfur dioxide"]<=60]
vino=vino[vino["total sulfur dioxide"]<=160]
vino=vino[vino["sulphates"]<=1.50]
vino=vino[vino["alcohol"]<=13.5]
```

```python
#Caracteristicas relevantes
#Condition
#El objetivo es predecir que personas tienen enfermedades cardiacas contenidas en la columna "condition"
print (vino["quality"].value_counts())
```

```python
columnas=["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol"]
predecir=["quality"]

columnas = vino[columnas]
predecir= vino[predecir]
```

```python
columnas.corr()
```

```python
seaborn.heatmap(columnas.corr())
pyplot.show()
```

```python
# Datos de prueba y entrenamiento
xentrenamiento, xprueba, yentrenamiento,yprueba= train_test_split(columnas,predecir,test_size=0.40,random_state=0)
print (xentrenamiento)
print (xprueba)
```

```python
clasificacion=DecisionTreeClassifier()
arbol=clasificacion.fit(xentrenamiento,yentrenamiento)
plot_tree(arbol)
```

```python
prediccionarbol=arbol.predict(xprueba)
prediccionarbol
```

```python
pandas.crosstab(pandas.array([y[0] for y in yprueba.values.tolist()]),prediccionarbol,rownames=["Actual"],colnames=["Prediccion Arbol"])
```

```python
#Evaluar el desempeo
print(classification_report(yprueba,prediccionarbol))
```
